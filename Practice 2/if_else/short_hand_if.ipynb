```python
number = 10

result = "Чётное" if number % 2 == 0 else "Нечётное"

print(result)
```

```python
age = 16

print("Взрослый" if age >= 18 else "Ребёнок")
```

```python
age = 20

if age >= 18: print("Взрослый")
```
