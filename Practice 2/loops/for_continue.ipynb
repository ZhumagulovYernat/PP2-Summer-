Что делает continue?

continue пропускает текущую итерацию и переходит к следующей.

```python
for letter in "Python":

    if letter == "h":
        continue

    print(letter)
```

```python
for i in range(10):

    if i % 2 == 0:
        continue

    print(i)
```

```python
for i in range(5):
    if i == 2:
        continue

    print(i)
```
