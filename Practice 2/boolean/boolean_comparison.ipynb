Оператор	Значение	Пример
==	равно	5 == 5
!=	не равно	5 != 3
>	больше	7 > 2
<	меньше	2 < 7
>=	больше или равно	5 >= 5
<=	меньше или равно	4 <= 5

```python
login = "admin"
password = "1234"

print(login == "admin")
print(password == "1234")
```

```python
score = 85

print(score >= 80 and score <= 100)
```

```python
x = 10
y = 20

print(x < y)
print(x > y)
print(x == y)
```

```python
print("apple" == "apple")
print("apple" == "banana")
```

```python
print(18 >= 18)
print(20 >= 18)
print(15 >= 18)
```

```python
print(5 < 8)
print(12 < 7)
```

```python
print(10 == 10)
print(10 == 5)
```
